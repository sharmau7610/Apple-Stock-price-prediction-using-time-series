# Time Series Analysis on Open, Closing, and Volume Stock
1. Prepare the Data

```python
open_stock = apple_data[['Open']]
close_stock = apple_data[['Close']]
volume_stock = apple_data[['Volume']]
```

2. create a function for plot

```python
def plot_timeseries(tdf,label1,label2,col,title):
    
    # Figuring the size
    plt.figure(figsize=(25,15))
    
    # Create a lineplot for data 1
    sns.lineplot(data=tdf,
                 label=label1,
                 legend=False,)
    
    # Create a lineplot for data 2
    sns.lineplot(data=apple_data[col],
                 label=label2,
                 legend=False,)
    
    # Create a title
    plt.title(title)
    
    plt.legend()
    
    # Show the plot
    plt.show()
```

# Simple Moving Average
A simple moving average (SMA) is an arithmetic moving average calculated by adding recent prices and then dividing that figure by the number of time periods in the calculation average.

1. Open Stock

```python
simple_moving_average_open_stock = open_stock.rolling(window=30).mean()

# Plotting Data
plot_timeseries(simple_moving_average_open_stock, # Data
                'Moving Average', # Label 1
                'Actual', # Label 2
                'Open', # Column of the Data
                'Open Stock Moving Average') 
```

2. Close Stock

```python
simple_moving_average_close_stock = close_stock.rolling(window=30).mean()

# Plotting Data
plot_timeseries(simple_moving_average_close_stock, # Data
                'Moving Average', # Label 1
                'Actual', # Label 2
                'Close', # Column of the Data
                'Close Stock Moving Average') #  
```

3. Volume Stock

```python
simple_moving_average_volume_stock = volume_stock.rolling(window=30).mean()

# Plotting Data
plot_timeseries(simple_moving_average_volume_stock, # Data
                'Moving Average', # Label 1
                'Actual', # Label 2
                'Volume', # Column of the Data
                'Volume Stock Moving Average') 
```

# Weighted Moving Average
A Weighted Moving Average puts more weight on recent data and less on past data. This is done by multiplying each bar's price by a weighting factor. Because of its unique calculation, WMA will follow prices more closely than a corresponding Simple Moving Average.



1. open stock

```python
weights = np.arange(1, 51)
open_MV = apple_data['Open'].rolling(50).apply(lambda close: np.dot(close, weights)/weights.sum(), 
                                      raw=True)

# Plotting Data
plot_timeseries(open_MV, # Data
                'Weighted Moving Average', # Label 1
                'Actual', # Label 2
                'Open', # Column of the Data
                'Open Stock Weighted Moving Average') # Title
```

2. close stock

```python
weights = np.arange(1, 51)
open_MV = apple_data['Close'].rolling(50).apply(lambda close: np.dot(close, weights)/weights.sum(), 
                                      raw=True)

# Plotting Data
plot_timeseries(open_MV, # Data
                'Weighted Moving Average', # Label 1
                'Actual', # Label 2
                'Close', # Column of the Data
                'Close Stock Weighted Moving Average') # Title
```

```python
weights = np.arange(1, 51)
volume_MV = apple_data['Volume'].rolling(50).apply(lambda close: np.dot(close, weights)/weights.sum(), 
                                          raw=True)

# Plotting Data
plot_timeseries(volume_MV, # Data
                'Weighted Moving Average', # Label 1
                'Actual', # Label 2
                'Volume', # Column of the Data
                'Closing Stock Weighted Moving Average')
```

# Exponential Moving Average
The exponential moving average (EMA) is a technical chart indicator that tracks the price of an investment (like a stock or commodity) over time. The EMA is a type of weighted moving average (WMA) that gives more weighting or importance to recent price data.

1. Open Stock

```python
expo_mv_open = apple_data['Open'].ewm(span=50, 
                              adjust=False).mean()

# Plotting Data
plot_timeseries(expo_mv_open, # Data
                'Exponential Moving Average', # Label 1
                'Actual', # Label 2
                'Open', # Column of the Data
                'Open Stock Exponential Moving Average') # 
```

2. Close Stock

```python
expo_mv_open = apple_data['Close'].ewm(span=50, 
                              adjust=False).mean()

# Plotting Data
plot_timeseries(expo_mv_open, # Data
                'Exponential Moving Average', # Label 1
                'Actual', # Label 2
                'Close', # Column of the Data
                'Close Stock Exponential Moving Average') # 
```

```python
expo_mv_volume = apple_data['Volume'].ewm(span=50, 
                                  adjust=False).mean()

# Plotting Data
plot_timeseries(expo_mv_volume, # Data
                'Exponential Moving Average', # Label 1
                'Actual', # Label 2
                'Volume', # Column of the Data
                'Volume Stock Exponential Moving Average')
```

# Time Series Lag Plot
A lag plot helps to check if a time series data set is random or not. A random data will be evenly spread whereas a shape or trend indicates the data is not random.
